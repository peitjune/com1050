Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là
độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh
cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân,
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn
hình

```python
import math
a, b = map(float, input("Nhap do dai 2 canh").split())
c = float(input("Nhap do dai canh huyen"))
p = (a + b + c) / 2
S = (math.sqrt(p*(p - a)*(p - b)*(p - c)))
if a+b>c and a+c>b and b+c>a:
    print("Day la do dai cua 1 tam giac")
else:
    print("Day khong phai do dai cua 1 tam giac")

if a**2 + b**2 == c**2:
    print("Day la tam giac vuong")
elif a == b and b == c:
    print("Day la tam giac deu")
elif a == b or a == c or b == c:
    print("Day la tam giac can")
else:
    print("Day la tam giac thuong")
print("Chu vi tam giac la",(a + b + c))
print(f"Dien tich tam giac la {S:.2f}")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a, b, c = map(float, input("Nhap 3 he so a b c").split())
delta = b**2 - 4*a*c
if delta > 0:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print(f"Phuong trinh co 2 nghiem phan biet x1 = {x1}, x2 = {x2}")
elif delta == 0:
    x = (-b) / (2*a)
    print(f"Phuong trinh co nghiem kep la {x}")
else:
    print("Phuong trinh vo nghiem")
```
