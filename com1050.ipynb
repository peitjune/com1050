3. mod and div
## / : chia thuc : 7/3 = 1.33333...
## // : chia lay phan nguyen : 7//3 = 2
## % : chia lay phan du : 7%3 = 1
## ** : phep mu : 2**3 = 8
## * : phep nhan : 2*3 = 6
<br> Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tổng, hiệu, tích, phần
nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập
phân) theo định dạng đúng của output

```python
# a = int(input("Nhap so a"))
# b = int(input("Nhap so b"))
a, b = map(int, input().split())
print(f"{a} + {b} = {a + b} \n{a - b} \n{a * b} \n{a // b} \n{a % b} \n{a / b:.2f}")
```

```python
a1, b1, c1, a2, b2, a3 = map(int, input().split())
tb = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10
print(f"Diem trung binh la {tb:1f}")
```
